```python
def show_imgs(img_list,gray_mode=None,titles=None,image_name=None):
    if gray_mode is None :
        print(f" please give the gray mode ")
        return 

    num = len(img_list)
    for i in range(num):
        plt.subplot(1,num,i+1)
        print(f'shape :{img_list[i].shape}')
        plt.imshow(img_list[i],cmap='gray' if gray_mode[i] else None)
        if titles is not None:
            plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    # if image_name is not None:
    #     plt.savefig(image_name)
    # else :
    #     plt.savefig('c.jpg')



def normalize(data):
    return (data-data.min())/(data.max()-data.min())


```
